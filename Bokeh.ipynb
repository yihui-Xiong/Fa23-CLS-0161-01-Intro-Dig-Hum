```python
import pandas as pd

# URL to the Iris dataset in CSV format
data_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

# Read the dataset from the URL directly
df = pd.read_csv(data_url)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral3
from bokeh.models import ColumnDataSource

# Prepare the output method for Google Colab
output_notebook()

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Map each species to a color
species_list = df['species'].unique().tolist()
color_map = factor_cmap('species', palette=Spectral3, factors=species_list)

# Create a new plot with a title and axis labels
p = figure(title="Iris Morphology", x_axis_label='Petal Width', y_axis_label='Petal Length')

# Add a circle renderer with a size, color, and legend option
p.circle(x='petal_width', y='petal_length', source=source, size=10, color=color_map, legend_group='species')

# Show the results
show(p)
```

```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Load Gapminder data
data_url = "https://raw.githubusercontent.com/jennybc/gapminder/master/inst/extdata/gapminder.tsv"
gapminder_data = pd.read_csv(data_url, sep='\t')

# Filter data for a specific country, e.g., Japan
japan_data = gapminder_data[gapminder_data.country == 'Japan']

# Create a ColumnDataSource
source = ColumnDataSource(japan_data)

# Prepare Bokeh plot
output_notebook()
p = figure(title="Life Expectancy Over Time in Japan", x_axis_label='Year', y_axis_label='Life Expectancy')
p.line('year', 'lifeExp', source=source)

# Show the plot
show(p)
```

```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource

# Load COVID-19 data
data_url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid_data = pd.read_csv(data_url)

# Filter data for a specific country, e.g., United States
us_data = covid_data[covid_data['location'] == 'United States']

# Convert date from string to datetime
us_data['date'] = pd.to_datetime(us_data['date'])

# Prepare the data source
source = ColumnDataSource(us_data)

# Prepare Bokeh plot
output_notebook()
p = figure(x_axis_type="datetime", title="COVID-19 Total Cases Over Time in the United States", x_axis_label='Date', y_axis_label='Total Cases')

# Add a line renderer
p.line('date', 'total_cases', source=source)

# Add Hover tool
hover = HoverTool(tooltips=[('Date', '@date{%F}'), ('Total Cases', '@total_cases')],
                  formatters={'@date': 'datetime'})
p.add_tools(hover)

# Show the plot
show(p)
```

```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Load automobile dataset
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
auto_data = pd.read_csv(data_url, header=None)

# Define column names (based on dataset's documentation)
auto_data.columns = ['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration', 'num_doors', 'body_style', 'drive_wheels', 'engine_location', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_type', 'num_cylinders', 'engine_size', 'fuel_system', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']

# Filter and clean data (remove missing values, etc.)
auto_data = auto_data[auto_data['horsepower'] != '?']
auto_data['horsepower'] = auto_data['horsepower'].astype(int)

# Create a ColumnDataSource
source = ColumnDataSource(auto_data)

# Prepare Bokeh plot
output_notebook()
p = figure(title="Horsepower vs. Price", x_axis_label='Horsepower', y_axis_label='Price')
p.scatter('horsepower', 'price', source=source)

# Show the plot
show(p)
```

```python
import yfinance as yf
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, DatetimeTickFormatter

# Fetch historical data for Apple Inc.
aapl = yf.Ticker("AAPL")
df = aapl.history(period="1y")  # Fetch one year of historical data

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Prepare the output
output_notebook()

# Create a figure with datetime x-axis
p = figure(title="AAPL Stock Closing Prices - Past Year", x_axis_type="datetime", x_axis_label='Date', y_axis_label='Closing Price (USD)')

# Add a line glyph
p.line(x='Date', y='Close', source=source, line_width=2)

# Customize datetime axis formatting
p.xaxis.formatter = DatetimeTickFormatter(days="%b %d", months="%b '%y", years="%Y")

# Show the plot
show(p)
```

```python
import pandas as pd
from forex_python.converter import CurrencyRates
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral10

# Initialize currency rates object
c = CurrencyRates()

# Base currency
base_currency = 'USD'

# List of currencies to compare
currencies = ['EUR', 'GBP', 'JPY', 'AUD', 'CAD', 'CHF', 'CNY', 'SEK', 'NZD', 'ZAR']

# Fetch exchange rates
exchange_rates = {currency: c.get_rate(base_currency, currency) for currency in currencies}

# Convert to DataFrame
df = pd.DataFrame(list(exchange_rates.items()), columns=['Currency', 'Rate'])

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Prepare the output
output_notebook()

# Create a figure
p = figure(x_range=df['Currency'], title=f"Exchange Rates Against {base_currency}", x_axis_label='Currency', y_axis_label='Rate')
p.vbar(x='Currency', top='Rate', source=source, width=0.7, color=factor_cmap('Currency', palette=Spectral10, factors=df['Currency']))

# Show the plot
show(p)
```

```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Load Earthquake Data
earthquake_url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.csv"
earthquake_df = pd.read_csv(earthquake_url)

# Convert DataFrame to ColumnDataSource
source = ColumnDataSource(earthquake_df)

# Prepare the output
output_notebook()

# Create a scatter plot for earthquake magnitudes
p = figure(title="Global Earthquakes - Magnitude vs Time", x_axis_label='Time', y_axis_label='Magnitude', x_axis_type='datetime')
p.scatter(x='time', y='mag', source=source, color='red')

# Show plot
show(p)
```

```python
index_start = text.find('Once meek,')
index_end = text.find('But I know I still have plenty of time.')
start = index_start if index_start != -1 else 0
end = index_end if index_end != -1 else len(text)

# Extract the relevant part of the text
extracted_text = text[start:end]
```

```python
tale_paras = extracted_text.split('\r\n\r\n')
author = ['olga tokarczuk'] * len(tale_paras)
title = ['drive'] * len(tale_paras)

tale_df = pd.DataFrame({'author': author, 'title': title, 'text': tale_paras})
```
